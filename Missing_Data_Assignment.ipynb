# Employee Name

```python
df['Employee_Name'].isnull().sum() * 100 / len(df['Employee_Name'])
```

```python
df['Employee_Name'].fillna(df['EmpID'],inplace=True) #another way is to fill it with unknown 
```

# Sex & DOB

```python
df['Sex'].fillna('unknown',inplace=True) 
```

```python
df['DOB'].fillna('unknown',inplace=True)
```

# Absences 

```python
df['Absences'].mode().round(2) # Using the mode to ensure fairness for all employees :) 
```

```python
df['Absences'].fillna(df['Absences'].mode()[0].round(2),inplace=True) # #The approach mode seems reasonable 
```

```python
df.isnull().sum()
```

# Salary

```python
df['Salary'].isnull().sum() * 100 / len(df['Salary'])
```

```python
df.Salary.plot(kind="hist", bins=20, title="Salary Distribution");
```

```python
positions=df['PositionID'].unique()
positions
```

```python
df['Salary'].fillna(df.groupby('PositionID')['Salary'].transform('mean'),inplace=True)
# fill the null values in the "Salary" column using the average salary based on unique "PositionID" values
```

```python
df.isnull().sum()
```

```python
df[df['Salary'].isna()==True]
```

```python
df['Salary'].fillna(df['Salary'].mean(),inplace=True) 
```

```python
df.isnull().sum()
```

# DateofTermination

```python
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'])
```

```python
df.sort_values(by='DateofTermination',inplace=True)
```

```python
#We will give all null 'DateofTermination' values the same value as that of the last person's termination date
#fill null values with the previous non-null value (forwardward fill) 

df.fillna(method='ffill', inplace=True)
```

# ManagerID

```python
# i think the best way is to fill it by mode 
df['ManagerID'].fillna(df['ManagerID'].mode(),inplace=True) 
```
